```python
import pandas as pd
from pycaret.clustering import *
import plotly.io as pio
pio.renderers.default = 'iframe'
```

```python
reviews = pd.read_csv("tripadvisor_review.csv")
reviews.drop("User ID", axis = 1, inplace = True)
reviews
```

```python
ClusteringParameters = setup(reviews)
```

```python
print("For Cluster = 3")
_3clustermodel = create_model('kmeans', num_clusters = 3)
_3clusterprediction = assign_model(_3clustermodel)

print("For Cluster = 4")
_4clustermodel = create_model('kmeans', num_clusters = 4)
_4clusterprediction = assign_model(_4clustermodel)

print("For Cluster = 5")
_5clustermodel = create_model('kmeans', num_clusters = 5)
_5clusterprediction = assign_model(_5clustermodel)
```

```python
_3clusterprediction.to_csv("C:/Users/MOKSH/OneDrive/Desktop/ML lab/clustering/kMeans/No_Data_Processing/3clusterResult.csv")
print("Result file save sucessfully!!")
```

```python
_4clusterprediction.to_csv("C:/Users/MOKSH/OneDrive/Desktop/ML lab/clustering/kMeans/No_Data_Processing/4clusterResult.csv")
print("Result file save sucessfully!!")
```

```python
_5clusterprediction.to_csv("C:/Users/MOKSH/OneDrive/Desktop/ML lab/clustering/kMeans/No_Data_Processing/5clusterResult.csv")
print("Result file save sucessfully!!")
```
